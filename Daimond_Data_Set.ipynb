```python
df.value_counts('cut')
```

```python
df['cut'].value_counts().plot(kind='bar',color=['pink','red','orange','blue','green'])
plt.grid(ls=':',c='k')
plt.xlabel('Quality',fontsize=25)
plt.ylabel('Values',fontsize=25)

plt.show()
```

```python
df['color'].value_counts()
```

```python
df['color'].value_counts().plot(kind='pie',
                                autopct="%0.2f%%",
                                explode=[0,0,0,0,0,0,0.5])
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['x'].value_counts().plot(kind='kde')
df['y'].value_counts().plot(kind='kde')
df['z'].value_counts().plot(kind='kde')
plt.grid(ls=':',c='k')
plt.legend()
plt.show()
```
